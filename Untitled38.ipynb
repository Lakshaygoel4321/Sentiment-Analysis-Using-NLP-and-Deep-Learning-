## **NLP techniques**

#### Making a preprocessing function and in that function i am converting the text in lower case ,removing the special symbol and  also removing the stopwords

```python

def preprocessing(text):

    # convert to lower
    text = text.lower()
    # removing the numerical and special symbol
    text = re.sub('[^a-zA-Z]',' ',text)
    # removing  the extra space
    text = re.sub('\s+[^a-zA-Z]\s+','',text)
    # removing the stop words
    pattern = re.compile(r'\b(' + '|'.join(stopwords.words('english')) + r')\b\s*')
    text = re.sub(pattern,'',text)

    return text

df['text'] = df['text'].apply(preprocessing)

```

## Perform the PorterStemmer

```python

ps = PorterStemmer()
def using_porter(text):
  return ' '.join([ps.stem(word) for word in text.split()])

df['text'] = df['text'].apply(using_porter)
```

# Now Start buliding the model

#### First Spliting the data

```python
train_data,test_data = train_test_split(df,test_size=0.2,random_state=12)
train_data
```

## ***Tokenizer***

```python

token = Tokenizer(num_words=5000)
# fit the train data
token.fit_on_texts(train_data['text'])

x_train = token.texts_to_sequences(train_data['text'])
x_test = token.texts_to_sequences(test_data['text'])

x_train
```

## Pad Sequences

```python

x_train_sequence = pad_sequences(x_train,maxlen=150)
x_test_sequence = pad_sequences(x_test,maxlen=150)

x_train_sequence
```

## Sperate the target column in the train_data and test_data

```python

y_train = train_data['sentiment']
y_test = test_data['sentiment']

y_train
```

## Checking the Shape of target and input feature

```python
y_test.shape
```

```python
x_test_sequence.shape
```

# Model

```python

model = Sequential()
# embedding and first layer
model.add(Embedding(input_dim=5000,output_dim=150,input_length=150))
model.add(LSTM(64,return_sequences=True))
model.add(Dropout(0.3))
# second layer
model.add(LSTM(128,return_sequences=True))
model.add(Dropout(0.3))
# third layer
model.add(LSTM(264,return_sequences=True))
model.add(Dropout(0.3))
# fourth layer
model.add(LSTM(512))
# output layer
model.add(Dense(4,activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

model.summary()
```

## Fitting the model

```python

model.fit(x_train_sequence,y_train,epochs=10,validation_data=(x_test_sequence,y_test))
```

## Predict on the test data

```python

prob = model.predict(x_test_sequence)
```

## Sample

```python
pred = np.argmax(prob,axis=1)
pred
```
